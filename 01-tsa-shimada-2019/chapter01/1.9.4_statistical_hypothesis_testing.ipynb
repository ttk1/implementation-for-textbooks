```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df_historical = pd.read_csv("../../data/chapter01/m_quote.csv")
```

```python
df_historical
```

```python
df_historical.USD.plot()
```

```python
df_historical.USD.plot.hist()
```

```python
# 1次階差
# 階差をとることで値が欠損する行が生じるためdropna()でその行を削除する
df_diff1 = df_historical.USD.diff().dropna()
# 1次階差系列のヒストグラムの表示
df_diff1.plot.hist()
```

```python
import scipy.stats as stats

# 1次階差系列のShapiro-Wilk検定
stats.shapiro(df_diff1.values)
```

```python
# 原系列のShapiro-Wilk検定
stats.shapiro(df_historical.USD.values)
```
